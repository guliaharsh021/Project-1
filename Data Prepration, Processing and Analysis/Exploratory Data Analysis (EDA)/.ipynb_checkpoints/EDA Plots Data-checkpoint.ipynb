# Exploratory Data Analysis (EDA) on Plots Data

```python
# Loading the required libraries 
import pandas as pd
import numpy as np 

# Importing the csv file 
plots_df = pd.read_csv('/Users/guliaharsh021/Downloads/DA Documents /Projects/Project 1/Data Prepration, Processing and Analysis/Data Exploration and Cleaning/Cleaned Data/plots_data_cleaned.csv')

# Printing the first few rows of the dataframe
plots_df.head()
```

```python
# Finding the total number of rows in the dataframe
plots_df.shape[0]
```

```python
# Selecting the required columns
plots_df = plots_df[['Property Type', 'City/Locality', 'Property Size (sqft)', 'Price (INR)', 'Price (per sqft)']].reset_index(drop=True)
plots_df.head()
```

### Summary Statistics:

```python
# Descriptive statistics for numerical columns
print(plots_df.describe())
```

```python
# Value counts for categorical column
print(plots_df['City/Locality'].value_counts())
```

### Distribution Plots:

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.boxplot(x = 'Property Size (sqft)', data = plots_df)
```

```python
# The Inter Quartile Range for the column `Property Size (sqft)`
q1 = plots_df['Property Size (sqft)'].quantile(0.25)
q3 = plots_df['Property Size (sqft)'].quantile(0.75)
iqr = q3 - q1

print(f"Interquartile Range (IQR) for Property Size (sqft): {iqr}")
```

```python
# the upper and lower bounds
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

print(f"Lower Bound for outliers: {lower_bound}")
print(f"Upper Bound for outliers: {upper_bound}")
```

```python
# how many outliers are there in the `Property Size (sqft)` column

outliers = plots_df[(plots_df['Property Size (sqft)'] < lower_bound) | (plots_df['Property Size (sqft)'] > upper_bound)]
num_outliers = outliers.shape[0]
print(f"Number of outliers in Property Size (sqft) column: {num_outliers}")
```

```python
# removing the outliers from the `Property Size (sqft)` column.
plots_df = plots_df[(plots_df['Property Size (sqft)'] >= lower_bound) & (plots_df['Property Size (sqft)'] <= upper_bound)]
```

```python
sns.histplot(x = 'Property Size (sqft)', data = plots_df)
```

### Price Analysis:

```python
# Analyzed the distribution of property prices.
sns.histplot(x = 'Price (INR)', data = plots_df)
```

```python
# Investigated the relationship between property size and price.
plt.figure(figsize=(10, 10))
sns.scatterplot(x = 'Price (INR)', y = 'Property Size (sqft)', data = plots_df)
plt.show()
```

### Correlation Analysis:

```python
# Created a correlation matrix to identify relationships between numerical features.

# Select numerical columns and append the encoded categorical columns
numerical_df = plots_df[['Property Size (sqft)', 'Price (INR)', 'Price (per sqft)']]

# Calculate the correlation matrix
correlation = numerical_df.corr()
print(correlation)
```

```python
# Visualized the correlation matrix using a heatmap.

# Plot the correlation matrix
plt.figure(figsize=(5, 3))
sns.heatmap(correlation, annot=False, cmap='coolwarm')
plt.show()
```

```python
# Finding the correlation between `Price` and all other numerical columns.
correlation2 = correlation.corr()['Price (INR)']

print("Correlation between Age and other numerical columns:")
print(correlation2)
```
