# Exploratory Data Analysis (EDA) on Houses Data

```python
# Loading the required libraries 
import pandas as pd
import numpy as np 

# Importing the csv file 
houses_df = pd.read_csv('/Users/guliaharsh021/Downloads/DA Documents /Projects/Project 1/Data Prepration, Processing and Analysis/Data Exploration and Cleaning/Cleaned Data/houses_data_cleaned.csv')

# Printing the first few rows of the dataframe
houses_df.head()
```

```python
# Finding the total number of rows in the dataframe
houses_df.shape[0]
```

```python
# Selecting the required columns
houses_df = houses_df[['Property Type', 'City/Locality', 'BHK', 'Property Size (sqft)', 'Furnishing', 'Price (INR)', 'Price (per sqft)']].reset_index(drop=True)
houses_df.head()
```

### Summary Statistics:

```python
# Descriptive statistics for numerical columns
print(houses_df.describe())
```

```python
# Value counts for categorical columns
print(houses_df['City/Locality'].value_counts())
print(houses_df['BHK'].value_counts())
print(houses_df['Furnishing'].value_counts())
```

### Distribution Plots:

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.boxplot(x = 'Property Size (sqft)', data = houses_df)
```

```python
# The Inter Quartile Range for the column `Property Size (sqft)`
q1 = houses_df['Property Size (sqft)'].quantile(0.25)
q3 = houses_df['Property Size (sqft)'].quantile(0.75)
iqr = q3 - q1

print(f"Interquartile Range (IQR) for Property Size (sqft): {iqr}")
```

```python
# the upper and lower bounds
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

print(f"Lower Bound for outliers: {lower_bound}")
print(f"Upper Bound for outliers: {upper_bound}")
```

```python
# how many outliers are there in the `Property Size (sqft)` column

outliers = houses_df[(houses_df['Property Size (sqft)'] < lower_bound) | (houses_df['Property Size (sqft)'] > upper_bound)]
num_outliers = outliers.shape[0]
print(f"Number of outliers in Property Size (sqft) column: {num_outliers}")
```

```python
# removing the outliers from the `Property Size (sqft)` column.
houses_df = houses_df[(houses_df['Property Size (sqft)'] >= lower_bound) & (houses_df['Property Size (sqft)'] <= upper_bound)]
```

```python
sns.histplot(x = 'Property Size (sqft)', data = houses_df)
```

### Price Analysis:

```python
# Analyzed the distribution of property prices.
sns.histplot(x = 'Price (INR)', data = houses_df)
```

```python
# Investigated the relationship between property size and price.
plt.figure(figsize=(10, 10))
sns.scatterplot(x = 'Price (INR)', y = 'Property Size (sqft)', data = houses_df)
plt.show()
```

### Furnishing Analysis:

```python
# Analyzed the distribution of properties based on furnishing status.
# Created bar plot to visualize the count of properties in each furnishing category.

plt.figure(figsize=(6, 6))
sns.countplot(x='Furnishing', data=houses_df)
plt.title('Distribution of Properties Based on Furnishing Status')
plt.xlabel('Furnishing Status')
plt.ylabel('Count of Properties')
plt.show()
```

### Feature Engineering:

#### To include furnishing details in the correlation analysis, using one-hot encoding to convert this categorical feature into numerical values.

```python
# One-hot encode categorical variables
houses_df = houses_df[['Furnishing', 'BHK', 'Property Size (sqft)', 'Price (INR)', 'Price (per sqft)']]
encoded_df = pd.get_dummies(houses_df, columns=['Furnishing'])
```

### Correlation Analysis:

```python
# Created a correlation matrix to identify relationships between numerical features.

# Select numerical columns and append the encoded categorical columns
numerical_df = encoded_df[['Property Size (sqft)', 'Price (INR)', 'Price (per sqft)', 'BHK']]
combined_df = pd.concat([numerical_df, encoded_df.drop(columns=['Property Size (sqft)', 'Price (INR)', 'Price (per sqft)', 'BHK'])], axis=1)

# Calculate the correlation matrix
correlation = combined_df.corr()
print(correlation)
```

```python
# Visualized the correlation matrix using a heatmap.

# Plot the correlation matrix
plt.figure(figsize=(10, 7))
sns.heatmap(correlation, annot=False, cmap='coolwarm')
plt.show()
```

```python
# Finding the correlation between `Price` and all other numerical columns.
correlation2 = correlation.corr()['Price (INR)']

print("Correlation between Age and other numerical columns:")
print(correlation2)
```
